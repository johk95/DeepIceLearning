```python
#low energy files (6 of them)
plt.clf()
fig = plt.figure(figsize=figsize(1.0))
for i, energ in enumerate(energies['11029']):
    axis = fig.add_subplot(321+i)
    sns.distplot(np.log10(energ),ax=axis)
    if i in (4,5):
        plt.xlabel(r'$\log_{10}(\mathrm{E}/\mathrm{GeV})$')
    else:
        axis.axes.xaxis.set_ticklabels([])
    if i in (0,2,4):
        plt.ylabel("Distribuiton density") #sets it on the current axis
    else:
        axis.axes.yaxis.set_ticklabels([])

#plt.show()
savefig("11029_energies_gridplot")
```

```python
#example code
plt.close('all')
fig = plt.figure()
from mpl_toolkits.axes_grid1 import Grid
grid = Grid(fig, rect=111, nrows_ncols=(3,2),
            axes_pad=0.25, label_mode='L',
            )
for i, ax in enumerate(grid):
    sns.distplot(np.log10(energies['11029'][i]),ax=ax)
    ax.title.set_visible(False)
plt.tight_layout()
```

```python
#high energy files (7 of them)
plt.clf()
fig = plt.figure(figsize=figsize(1.0))
for i, energ in enumerate(energies['11069']):
    axis = fig.add_subplot(421+i)
    sns.distplot(np.log10(energ),ax=axis)
    if i in (5,6):
        plt.xlabel(r'$\log_{10}(\mathrm{E}/\mathrm{GeV})$')
    else:
        axis.axes.xaxis.set_ticklabels([])
    if i in (0,2,4,6):
        plt.ylabel("Distribuiton density") #sets it on the current axis
    else:
        axis.axes.yaxis.set_ticklabels([])

#plt.show()
savefig("11029_energies_gridplot")
```
